```python
import numpy as np
import pandas as pd

# Assuming df is your DataFrame
df = pd.read_csv('/content/bank-additional-full.csv',delimiter=';')
column = df[['age', 'campaign', 'duration']]
q1 = np.quantile(column, 0.25)
q3 = np.quantile(column, 0.75)  # Corrected from 0.25 to 0.75 for q3
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)  # Corrected from subtraction to addition for upper bound
print(iqr, upper_bound, lower_bound)

# Filtering out outliers
df[['age', 'campaign', 'duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
df.plot(kind='box',subplots=True,layout=(2,5),figsize=(20,10),color='#808000')
```
